```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'F:'
df = pd.read_csv(path + 'titanic.csv')
```

```python
df.head()
```

```python
inputs = df.drop(['Survived','Pclass','Parch', 'Ticket','Fare','Cabin','SibSp','Embarked'], axis = 'columns')
```

```python
inputs
```

```python
target = df['Survived']
```

```python
target
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le_pid = LabelEncoder()
le_name = LabelEncoder()
le_sex = LabelEncoder()
le_age = LabelEncoder()
```

```python
inputs['pid_n'] = le_pid.fit_transform(inputs['PassengerId'])
inputs['name_n'] = le_name.fit_transform(inputs['Name'])
inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs['age_n'] = le_age.fit_transform(inputs['Age'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['PassengerId', 'Name', 'Sex', 'Age'], axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```
